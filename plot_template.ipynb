```python
import pennylane as qml
import tensorflow as tf
import os
import pickle


import sklearn.datasets
from sklearn.datasets import fetch_openml

import numpy as np

from matplotlib import pyplot as plt

tf.keras.backend.set_floatx('float64')
```

```python
#function to save training history as pickle
def save_training_history(history, filename):
    history_path = os.path.join('history', filename)
    with open(history_path, 'wb') as f:
        pickle.dump(history.history, f, pickle.HIGHEST_PROTOCOL)
```

```python
#function to load training history pickles and plot the accuracy
def plot_hinge_accuracy(histories):
    fig, ax = plt.subplots(figsize=(12, 6))
    for history in histories:
        with open(os.path.join('history', history), 'rb') as f:
            history_dict = pickle.load(f)
        hinge_acc = history_dict['hinge_accuracy']
        epochs = range(1, len(hinge_acc) + 1)
        ax.plot(epochs, hinge_acc, label=history)
    ax.set_title('Accuracy for different Gradient Methods')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Hinge Accuracy')
    legend_elements = [
      plt.Line2D([0], [0], marker='o', color='#1f77b4', label='finite-diff'),
        plt.Line2D([0], [0], marker='o', color='#ff7f0e', label='parameter-shift'),
     plt.Line2D([0], [0], marker='o', color='#2ca02c', label='SPSA'),
       
         
                   
                   
    
]
    ax.legend(handles=legend_elements)
    plt.savefig("Gradient_Amp_Encoding_Accuracy.png", bbox_inches="tight", pad_inches=0)
    plt.show()
```

```python
#function to load training history pickles and plot the loss
def plot_loss(histories):
    fig, ax = plt.subplots(figsize=(12, 6))
    for history in histories:
        with open(os.path.join('history', history), 'rb') as f:
            history_dict = pickle.load(f)
        hinge_acc = history_dict['loss']
        epochs = range(1, len(hinge_acc) + 1)
        ax.plot(epochs, hinge_acc, label=history)
    ax.set_title('Loss for different Gradient Methods')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    legend_elements = [
   plt.Line2D([0], [0], marker='o', color='#1f77b4', label='finite-diff'),
        plt.Line2D([0], [0], marker='o', color='#ff7f0e', label='parameter-shift'),
     plt.Line2D([0], [0], marker='o', color='#2ca02c', label='SPSA'),
]
    ax.legend(handles=legend_elements)
    plt.savefig("Gradient_Amp_Encoding_Loss.png", bbox_inches="tight", pad_inches=0)
    plt.show()
```

```python
#Example to use the function
histories = ['amplitude_50_1hl_adam_finite_diff.pickle','amplitude_50_1hl_adam_parameter_shift.pickle','amplitude_50_1hl_adam_spsa.pickle']
plot_hinge_accuracy(histories)
plot_loss(histories)
```
